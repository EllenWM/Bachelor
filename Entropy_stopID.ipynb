```python
#number of stops at each stopID for each user
locations = data.groupby(['userId','stopID']).count()['incrementalId'].reset_index()
```

```python
locations
```

```python
#dataframe for the 3 entropies
entropy = pd.DataFrame()
```

```python
#random entropy
N = np.zeros(len(user))

for i in user:
    N[user.index(i)] = len(locations[locations.userId == i]) #number of distinct locations
```

```python
#random entropy
s_rand = np.log2(N)
```

```python
#random entropy
entropy['s_rand'] = s_rand
```

```python
entropy
```

```python
sUnc = np.zeros(len(user))
for i in user: #loop through users
    location = locations[locations.userId == i] #df with stops at stopID for user i
    probdf = pd.DataFrame(data=location)
    probdf['prob'] = location['incrementalId'] / location.sum()['incrementalId'] #prob added to the df for user i
    for j in range(0,len(location)): #loop through the locations for each user
        sUnc[user.index(i)] = sUnc[user.index(i)] + probdf['prob'].iloc[j] * np.log2(probdf['prob'].iloc[j])
s_unc = -sUnc
```

```python
s_unc
```

```python
#temporal-uncorrelated entropy
entropy['s_unc'] = s_unc
```

```python
#from all stops
entropy
```

```python
#stopID as the category
data['stopID'] = data['stopID'].astype('category')
```

```python
#label encoding
data['Stop_Cat'] = data['stopID'].cat.codes
```

```python
def contains(small, big):
    try:
        big.tostring().index(small.tostring())//big.itemsize
        return True
    except ValueError:
        return False

def actual_entropy(l):
    n = len(l)
    sum_lambda = 0

    for i in range(0, n):
        sequence = l[:i]
        
        for j in range(i+1, n+1):
            s = l[i:j]
            if contains(s, sequence) != True:
                sum_lambda += len(s)
                break
    
    h = (n*math.log2(n))/(sum_lambda)
    return h
```

```python
actualEntropy = np.zeros(len(user))
for u in user:
    userData = data[data.userId == u]['Stop_Cat'].values
    actualEntropy[user.index(u)] = actual_entropy(userData)
```

```python
actualEntropy
```

```python
entropy['s_actual'] = actualEntropy
```

```python
#relative difference
entropy['diff'] = (entropy['s_actual']-entropy['s_rand'])/entropy['s_rand']
```
