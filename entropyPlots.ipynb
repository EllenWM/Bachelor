```python
import os
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [20, 10]
import numpy as np
sns.set_theme()
sns.set_style("whitegrid")
sns.set_context("poster", font_scale = 1.2)
```

```python
#load of entropy files
os.chdir("/Users/Ellen/Documents/Dokumenter/DTU/7. Semester/Bachelor/CSV files")

entropy_cat = pd.read_csv("entropyBigram.csv")
entropy_ID = pd.read_csv("entropy_stopID.csv")
```

```python
colors = ["#C54E57", "#65ab7c", "#6495ED"] #dull red, tea, cornflower blue
sns.set_palette(sns.color_palette(colors))
```

```python
os.chdir("/Users/Ellen/Documents/Dokumenter/DTU/7. Semester/Bachelor/Plots")

sns.histplot(data=entropy_cat[['s_rand', 's_unc', 's_actual']])
plt.xlabel('Entropy')
plt.ylabel('Number of users')
plt.legend(loc='upper right', labels=['rand entropy', 'unc entropy', 'actual entropy'])
plt.savefig('EntropyHist_stopCat.png', bbox_inches='tight') #Save plot
plt.show()
```

```python
sns.histplot(data=entropy_ID[['s_rand', 's_unc', 's_actual']])
plt.xlabel('Entropy')
plt.ylabel('Number of users')
plt.legend(loc='upper right', labels=['rand entropy', 'unc entropy', 'actual entropy'])
plt.savefig('EntropyHist_stopID.png', bbox_inches='tight') #Save plot
plt.show()
```

```python
sns.histplot(data=entropy_cat['s_rand'])
sns.histplot(data=entropy_ID['s_rand'],color = 'cornflowerblue')
plt.legend(loc='upper left', labels=['stop type', 'stop ID'])
plt.xlabel('Entropy')
plt.ylabel('Number of users')
plt.savefig('randomEntropy.png', bbox_inches='tight')
plt.show()
```
